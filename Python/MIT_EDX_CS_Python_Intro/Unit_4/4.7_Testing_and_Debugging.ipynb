# Section 4.7: Testing and Debugging - 11/16

*<ins>Defensive Programming</ins>*
* Write specifications
* Modularize programs
* Check conditions on inputs/outputs (assertions)

*<ins>Testing/Validation</ins>*
* Compare input/output pairs to specification
* "its not working!"
* "How can i break my program?"

*<ins>Debugging</ins>*
* Find and fix errors
* "Why isn't it working?"
* "How can i fix my program?"
* Study events leading to error





#### <ins>Classes of Tests</ins>

*<ins>Set Yourself Up for Easy Testing and Debugging</ins>*
* From the start, write code that is easy to test and debug
* Write modular code
* Document constraints on modules
    * What do you expect the inputs to be? What do you expect the outputs to be?
* Document assumptions behind code design

*<ins>When are you ready to test?</ins>*
* When you have a working program
    * Remove syntax errors
    * Remove statis semantic errors
    * Python can usually find these for you
* Have a set of expected results
    * What should the program do?
    * What should the output be?

*<ins>Classes of Tests</ins>*
* Unit Testing
    * Validate each piece of program
    * Test each function separately

* Regression Testing
    * Add test for bugs as you find them in a function 
    * catch reintroduced errors that were previously fixed

* Integration Testing
    * Does overall program work?
    * Tend to rush to do this

*<ins>Testing Approaches</ins>*
* Intuition about natural bounderies to the problem
    * Can you come up with some natural partitions? 
* If no natural partitions, might do random testing 
    * probability that code is correct increases with more tests 
    * Better options below 
* **Black Box** testing
    * Explore paths through specification 
    * Designed without looking at the code 
    * Can be done by someone other than the implementer to avoid some implementer biases 
    * Testing can be reused if implementation changes
    * Paths through specification
        * build test cases in different natural space partitions 
        * also consider boundary conditions
* **Glass Box** testing
    * Explore paths through code
    * Use code directly to guide design of test cases 
    * Called path-complete if every potential path through code is tested at least once 
    * What are some drawbacks
        * can go through loops arbitrarily many times 
        * Missing paths 
    * guidelines:
        * branches 
        * for loops
        * while loops



#### <ins>Bugs</ins>
* Once you have discovered a bug, you need to find it and fix it. Retest it to make sure it is fixed.
**Overy vs. Covert**
* Overt: Crash, wrong answer, etc.
* Covert: Program runs to completion, but produces wrong answer

**Persistent vs. Intermittent**
* Persistent: Occurs every time you run the program
* Intermittent: Occurs only sometimes

#### <ins>Debugging</ins>
* Steep learining curve
* Goal is to have a bug-free program
* tools:    
    * Buit in to IDLE and Anaconda
    * print statements
    * Python Tutor
    * Be systmatics 

*<ins>Logic Errors - Hard to Find</ins>*
* Think before writing code
* Draw pictures
* Use print statements
* Explain the code

*<ins>Debugging Steps</ins>*
* Study program code
    * Make sure you understand it
    * Ask how did I get the unexpected result
    * Don't ask what is wrong
    * Is it part of a family

* Scientific Method
    * Hypothesis
    * Tests
    * Repeat
    * Pick simplest input to test with



#### <ins>Debugging Skills</ins>
* Treat as a search problem: Looking for explanation for incorrect behavior
    * Study data
    * Form a hpyothesis
    * Design and run experiments to test hypothesis
    * Keep records of experiments

*<ins>Debugging as Search</ins>*
* Want to narrow down space of possible sources of error 
* Design experiment that expose intermediate stages of computation 
* Binary search can be helpful
